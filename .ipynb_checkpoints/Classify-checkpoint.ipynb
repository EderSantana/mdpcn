```python
cost = ConvSparseReconstructionError()
dataset = VIDTIMIT('train', axes=('b','c',0,1), center=True)
#monitoring_dataset = {'valid': VIDTIMIT('test', axes=('b','c',0,1), center=True, noise_v=.5)}
monitoring_dataset = {'valid': VIDTIMIT('train', axes=('b','c',0,1), center=True, noise_v=.5)}
model = cPickle.load(file('model_small.pkl'))
trainer = Trainer(cost, model, dataset, monitoring_dataset, .06, './results.pkl',2)
n_pp = 10
n_ex = 8
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as KNC
from sklearn.utils import shuffle
from sklearn.metrics import precision_score
#testdata  = monitoring_dataset['valid'].X.reshape(72,54,32,32)
testdata  = monitoring_dataset['valid'].X.reshape(288,54,32,32)
traindata = dataset.X.reshape(288,54,32,32)

print testdata.shape
print traindata.shape
```

```python
train1 = np.zeros((288, 54, 16, 36, 36)).astype('float32')
train2 = np.zeros((288, 54, 16, 22, 22)).astype('float32')
for i in range(288/8):
    print i
    for t in range(54):
        if t==0:
            flag = 0
        else:
            flag = 1
        start = i*8
        end   = (i+1)*8
        model.layers[0].S.set_value(traindata[start:end,t:t+1,:,:])
        model.layers[0].update_top_state(model.layers[1].get_reconstruction() * flag)
        #model.layers[0].update_top_state(model.layers[0].X.get_value() * flag)
        model.layers[0].opt.run( model.layers[0].n_steps )
        train1[start:end,t,:,:,:] = model.layers[0].X.get_value()
        out = trainer.compiled_fprop[0](traindata[start:end,t:t+1,:,:])
        
        model.layers[1].S.set_value(out)
        model.layers[1].update_top_state(model.layers[1].X.get_value() * flag)
        model.layers[1].opt.run( model.layers[0].n_steps )
        train2[start:end,t,:,:,:] = model.layers[1].X.get_value()
        
        
```

```python
#layer1 = np.zeros((72, 54, 16, 36, 36)).astype('float32')
layer1 = np.zeros((288, 54, 16, 36, 36)).astype('float32')
#layer2 = np.zeros((72, 54, 16, 22, 22)).astype('float32')
layer2 = np.zeros((288, 54, 16, 22, 22)).astype('float32')
for i in range(288/8):
    print i
    for t in range(54):
        if t==0:
            flag = 0
        else:
            flag = 1
        start = i*8
        end   = (i+1)*8
        model.layers[0].S.set_value(testdata[start:end,t:t+1,:,:])
        model.layers[0].update_top_state(model.layers[1].get_reconstruction() * flag)
        #model.layers[0].update_top_state(model.layers[0].X.get_value() * flag)
        model.layers[0].opt.run( model.layers[0].n_steps )
        layer1[start:end,t,:,:,:] = model.layers[0].X.get_value()
        out = trainer.compiled_fprop[0](testdata[start:end,t:t+1,:,:])
        
        model.layers[1].S.set_value(out)
        model.layers[1].update_top_state(model.layers[1].X.get_value() * flag)
        model.layers[1].opt.run( model.layers[0].n_steps )
        layer2[start:end,t,:,:,:] = model.layers[1].X.get_value()
        
```

```python
train1 = train1.reshape((288*54, -1))
train2 = train2.reshape((288*54, -1))
layer1 = layer1.reshape((288*54, -1))
#layer2 = layer2.reshape((72*54, -1))
layer2 = layer2.reshape((288*54, -1))
print layer1.shape
#print layer2.shape
```

```python
train_label = dataset.get_targets()
test_label = monitoring_dataset['valid'].get_targets()
ex_label = np.zeros((288*54))
ex_test = np.zeros((288*54))

for i in range(288):
    start = i*54
    end = (i+1)*54
    ex_label[start:end] = train_label[i]
for i in range(288):
    start = i*54
    end = (i+1)*54
    ex_test[start:end] = test_label[i]
```

### Layer 1

```python
from sklearn.decomposition import PCA, KernelPCA
from sklearn.svm import SVC
from sklearn.utils import shuffle
from sklearn.metrics import precision_score
```

```python
#Xt, Yt = shuffle(Y_l1, ex_label)
#clf = SVC(kernel='rbf')
pca1 = KernelPCA(kernel='linear', gamma=.01, n_components=20)

data_train = train1
t_train = ex_label[:54*8*7]
data_test  = layer1
t_test = ex_test[:54*8*7]
Y = pca1.fit_transform( data_train[:54*8*7]  )
T = pca1.transform( data_test[:54*8*7] )

clf = KNC(n_neighbors=22)
#clf = SVC(kernel='rbf',gamma=.01)
clf.fit(Y, t_train)
Ytrain = clf.predict(Y)
Ytest = clf.predict(T)
print precision_score(t_train, Ytrain)
print precision_score(t_test,Ytest)
```

```python
print Ytrain.shape
print Ytest.shape
```

```python
from scipy.stats.mstats import mode
def my_mode(S):
    Y = np.zeros(S.shape[0])
    for i in range(S.shape[0]):
        Y[i] = mode(S[i])[0]
    return Y
```

```python
Ytest_new = my_mode(Ytest.reshape(-1,54))
Ytrain_new = my_mode(Ytrain.reshape(-1,54))
print precision_score(train_label[:8*7], Ytrain_new)
print precision_score(test_label[:8*7], Ytest_new)
print (test_label[:7*8]==Ytest_new).sum()
```
